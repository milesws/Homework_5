```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import collections
from sklearn.datasets import make_regression
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
```

```python
from sklearn.datasets import fetch_mldata

mnist = fetch_mldata("MNIST original")
X = mnist.data / 255.0
y = mnist.target
print(X.shape)
print(y.shape)
collections.Counter(y)
```

```python
dataframe = pd.DataFrame(data=X, index=y)
dataframe.head()
```

## **Part One**

First I will see how much variance is explained with 2 components

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(X)
pca.explained_variance_ratio_
```

Because the amount of variance explained is fairly low, I'll plot the ratio cumulatively for all components

```python
pca = PCA(n_components=100)
data_pca = pca.fit_transform(X)
cum_ratio = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(len(cum_ratio)), cum_ratio)
plt.vlines(ymin=0,ymax=cum_ratio[20],x=20)
```

It seems that the marginal benefit (percent of variance explained) decreases at 20 components, so I'll use 20.

```python
plt.xlabel('Comp1')
plt.ylabel('Comp2')
comp1 = data_pca[0]
comp2 = data_pca[1]
sns.regplot(comp1, comp2, fit_reg=False)
```

There seems to be a positive trend between comp 1 and comp 2.  This shows that both have positive slopes in explaining more variance.
